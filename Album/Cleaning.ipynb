```python
from datascience import *
import numpy as np
```

```python
albums = Table.read_table("album_classifications.csv")
albums
```

```python
grouped_type = albums.group("Type").sort("count", descending = True)
grouped_type.show()
```

```python
albums.where("Type", "Music")
```

```python
double = albums.group(["Artist", "Type"])
double
```

```python
single = albums.group("Artist")
single
```

```python
percentages = double.join("Artist", single)
percentages.relabel("count", "Releases")
percentages.relabel("count_2", "# Albums in Catalog")
percentage = percentages.column("Releases") / percentages.column("# Albums in Catalog")
percentages = percentages.with_column("Percentage of Catalog", percentage)
percentages.show()
```

```python
percentages.to_csv("catalog_percentages.csv")
```
